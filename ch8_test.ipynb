```python
import ipywidgets as widgets
from IPython.display import display

print("Hello StreamLit")

name_input = widgets.Text(value='', placeholder='Enter your Name', description='Name:', disabled = False)
output = widgets.Output()

def on_name_change(change):
    with output:
        output.clear_output()
        print(f"Hello {change['new']}")

name_input.observe(on_name_change, names='value')
display(name_input, output)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

print("Number Input")

n1 = widgets.IntText(value = 0, description='First Number:')
n2 = widgets.IntText(value = 0, description='Second Number:')
output = widgets.Output()

def calculate_sum(change):
    with output:
        clear_output()
        sum_result = n1.value + n2.value
        print(f"The sum is : {sum_result}")

n1.observe(calculate_sum, names='value')
n2.observe(calculate_sum, names='value')
display(n1, n2, output)
calculate_sum(None)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

def colorIndex(colorData, findColor="紅色"):
    for i, color in enumerate(colorData):
        if color == findColor:
            return i
    return -1

data = ["紅色", "藍色", "綠色"]
dropdown = widgets.Dropdown(options = data, description='你喜歡甚麼顏色：', value='紅色')
output = widgets.Output()

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        with output:
            clear_output()
            selected_color = change['new']
            idx = colorIndex(data, selected_color)
            print(f"你選擇的顏色是：{selected_color}，索引為：{idx}")

dropdown.observe(on_change)
display(dropdown, output)
with output:
    selected_color = dropdown.value
    idx = colorIndex(data, selected_color)
    print(f"你選擇的顏色是：{selected_color}，索引為：{idx}")
```
